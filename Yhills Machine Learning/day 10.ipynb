```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Loading the data
iris = datasets.load_iris()
X = iris.data  # Data details
Y = iris.target  # Labels for data

# Splitting the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

# Standardizing the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# SVM model
svm_model = SVC(kernel='linear')
svm_model.fit(X_train, Y_train)
Y_pred = svm_model.predict(X_test)

# Finding the accuracy
accuracy = accuracy_score(Y_test, Y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')

# Visualizing the data
X_vis = iris.data[:, :2]  # Taking only the first two features for visualization
Y_vis = iris.target

X_train_vis, X_test_vis, Y_train_vis, Y_test_vis = train_test_split(X_vis, Y_vis, test_size=0.3, random_state=42)

X_train_vis = scaler.fit_transform(X_train_vis)
X_test_vis = scaler.transform(X_test_vis)

svm_model.fit(X_train_vis, Y_train_vis)

x_min, x_max = X_vis[:, 0].min() - 1, X_vis[:, 0].max() + 1
y_min, y_max = X_vis[:, 1].min() - 1, X_vis[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))
Z = svm_model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plotting the graph
plt.contourf(xx, yy, Z, alpha=0.8)
plt.scatter(X_vis[:, 0], X_vis[:, 1], c=Y_vis, edgecolors='k', marker='o')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('SVM Decision Boundary')
plt.show()
```

```python
#handling the missing data
import pandas as pd
from sklearn.impute import SimpleImputer

data={"A":[1,2,None,4],"B":[None,2,3,4],"C":[5,6,7,8]}
df= pd.DataFrame(data)
#removing the null data init
df_dropna=df.dropna()
#then putiing back missing value
imputer= SimpleImputer(strategy="mean") #replace the missing value with medium value or mean
df_imputed= pd.DataFrame(imputer.fit_transform(df),columns=df.columns)

print("Original data")
print(df)
print("droped data")
print(df_dropna)
print("Filled data")
print(df_imputed)
```

```python
#standardization of data
from sklearn.preprocessing import StandardScaler
import pandas as pd


data={"A":[1,2,None,4],"B":[None,2,3,4],"C":[5,6,7,8]}
df= pd.DataFrame(data)

scaler = StandardScaler()
df_scaled =pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
print("\n Original")
print(df)
print("\nStandardization")
print(df_scaled)
```

```python
#Normalization of data
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

data={"A":[1,2,3,4],"B":[2,4,5,6]}
df= pd.DataFrame(data)

scaler=MinMaxScaler()
df_normalized= pd.DataFrame(scaler.fit_transform(df),columns=df.columns)

print("\n orginal")
print(df)
print("new data")
print(df_normalized)
```
