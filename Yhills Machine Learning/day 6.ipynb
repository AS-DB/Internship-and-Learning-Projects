Reading CSV file

```python
#csv file accesing data using panda
import pandas as pd

#declaring a varable to to read csv file

reader = pd.read_csv('tips.csv')
#showing few data 
print(reader)

#showing whole data
print(reader.to_string)

#to show particular data from start using .head() which will return 5 data
print(reader.head())

#to show particular number data from start using .head(int)
print(reader.head(10))

#to show particular data from bottom using .tail() which will return 5 data
print(reader.tail())

#to show particular number data from bottom using .tail(int)
print(reader.tail(10))
```

Reading Json file

```python
#reading json file
import pandas

reader= pd.read_json('anscombe.json')
print(reader)

#reading starting 5 elements using .head()
r1 =reader.head()
print(r1)


#reading starting 10 elements using .head(10)
r2 =reader.head(10)
print(r2)

#reading bottom 5 elements using .tail()
r3= reader.tail()
print(r3)

#reading ending  10 elements using .tail(10)
r4 =reader.tail(10)
print(r4)
```

DATA Cleaning

```python
#Data cleaing 
import pandas as pd
reader = pd.read_csv("data_cleaing.csv")

#removing null value in data set
reader_new = reader.dropna()
print(reader_new)

#filling empty value in data set which wil makes parent change to data in main file
reader_new1 = reader.fillna(130,inplace=True)
print(reader_new1)

#filling empty value in data set which wil make any change in main file
reader_new2 = reader.fillna(130)
print(reader_new2)

#removing the duplicates data from in csv file in temporary
reader_new3 = reader.drop_duplicates()
print(reader_new3)


#removing the duplicates data from in csv file in paramanently
reader_new4 = reader.drop_duplicates(inplace=True)
print(reader_new4)

#to check wheter its has duplicate value in set if it has it will return duplicate value or return false

reader_new5 = reader.duplicated()
print(reader_new5)
```
