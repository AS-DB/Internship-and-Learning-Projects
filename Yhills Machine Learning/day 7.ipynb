Matplotlib module:-

Its a low level graph plotting library that serves as visualization utility.

Matplotlib functions:-
    1.Pyplot

    Pyplot Matplotlib utilities lies under the pyplot submodule which we can and we can define plot a graph using it

    ->In order to work with it we use to import both matplotlib.pyplot and  panda library

    ->then declare cordinates of graph using numpy array
        
        Coordinate_varable= np.array([int,int,int...])

    ->Then we .plot function in order to assign axis in graph
        
        matplotlib.pyplot.pyplot(coordinate_variables)
    
    ->Then to show it we use:

        matplotlib.pyplot.show()

    2.Marker(Points on graph):-These are plot points are used to show points on graph which have their own size,design,color
        -> To change design we use
            matplotlib.pyplot.plot(coordinates,marker="design-shape")
        
        ->changing design  of line and its color without marker include
            matplotlib.pyplot.plot(cordinates_varaible,"line_designcolor->color") e.g. matplotlib.pyplot.plt(x,y,"-r",marker="*")
       
        ->changing design  of line and its color with marker include
            matplotlib.plot(cordinates_varaible,"marker->line_design->color") e.g. matplotlib.pyplot.plt(x,y,"*-r")

        ->changing the size of marker
        matplotlib.pyplot.plot(cordinates_varaible,ms=font-size) e.g. matplotlib.pyplot.plt(x,y,ms=21)
        
        matplotlib.pyplot.plot(cordinates_varaible,markersize=font-size) e.g. matplotlib.pyplot.plt(x,y,markersize=21)

        ->changing the face-color of marker
        
        matplotlib.pyplot.plot(cordinates_varaible,mfc="color") e.g. matplotlib.pyplot.plt(x,y,mfc="red")

        ->changing the edge color of marker

        matplotlib.pyplot.plot(cordinates_varaible,mec="color") e.g. matplotlib.pyplot.plt(x,y,mec="red")

        ->Changing the line design

        matplotlib.pyplot.plot(cordinates_varaible,linestyle="line-style name") e.g. matplotlib.pyplot.plt(x,y,linestyle="dotted")

        ->Changing the line design width

        matplotlib.pyplot.plot(cordinates_varaible,linewidth=widthsize) e.g. matplotlib.pyplot.plt(x,y,linewidth=widthsize)

    3.Multiple line in single plot
        
        variablename_1=numpy.array([1,3,4])
        variablename_2=numpy.array([6,7,4])

        matplotlib.pyplot.show()

    4.Giving labels in x-axis and y-axis
        
    ->x-axis:-
        matplotlib.pyplot.xlable("Label_name")
        
    ->y-axis:-
        matplotlib.pyplot.ylable("Label_name")

    5.Giving Title graph and its location to change

        matplotlib.pyplt.title("title-name")
        matplotlib.pyplt.title("title-name",loc="direction")
    
    6.Giving grids in graph x axis or y-axis or both axis

        matplotlib.pyplt.grid(axis='x')
                    or
        matplotlib.pyplt.grid(axis='y')
                    or
        matplotlib.pyplt.grid(axis='x',axis='y')

        then use .show():-
        matplotlib.pyplt.show()
        
    7.changing the color and linesize of grid
        
        matplotlib.pyplot.grid(color="color",linestyle="linestyle-name",linewidth=font-width)

        e.g.
        plt.grid(color="red",linestyle="dotted",linewidth=5)

    8.How to plot multiple graph in single graph set 

            matplotlib.pyplot.subplot(int,int,int) #row,column,plotnumber
    
    9.Scatter-graph

            matplotlib.pyplot.scatter(Cordinates_varables)
            matplotlib.pyplot.show()
                    
    10.Changing color of scattered point in scattered graph

            matplotlib.pyplot.scatter(Cordinates_varables,color="color")

            e.g
            matplotlib.pyplot.scatter(Cordinates_varables,color="red")
    
    11.Bar-graph

        matplotlib.pyplot.bar(Cordinates_varables)
        matplotlib.pyplot.show()
    
    12.Horizontal bar graph

        matplotlib.pyplot.barh(Cordinates_varables)
        matplotlib.pyplot.show()
    
    13.Change the color,width-size of graph

        matplotlib.pyplot.bar(coordinates,color,width)
        matplotlib.pyplot.show()

```python
import matplotlib.pyplot as plt
import numpy as np
#declarding x-y graph
x_point = np.array([0,6])
y_point = np.array([50,20])
#plotting the coordinates
plt.plot(x_point,y_point)
#use to display the plotted graph
plt.show()
```

```python
#using marker in graph to highlight the point
import matplotlib.pyplot as plt
import numpy as np
x_point = np.array([0,6])
y_point = np.array([50,20])
#plotting the coordinates using marker in parenthesis
plt.plot(x_point,y_point,marker ='*')
plt.show()
```

```python
#changing the colour in graph of line and its vertex color
import matplotlib.pyplot as mp
import numpy as np
y_point=np.array([2,3,4,5])
#changing the color with line design 
mp.plot(y_point,"--r",marker="o") #here -r equals to - equal to line design of graph and r equal to red
                                # we can directly write the colur name
mp.show()

mp.plot(y_point,'*-r') # *->marker design - ->line design r ->red
mp.show()
```

```python
#changing the colour in graph of line and its vertex color
import matplotlib.pyplot as mp
import numpy as np
y_point=np.array([2,3,4,5])
x_point=np.array([6,7,8,9])
#changing the color with line design 
mp.plot(y_point,x_point,'*-r') # *->marker design - ->line design r ->red
mp.show()
```

```python
#changing the size of marker
import matplotlib.pyplot as mp
import numpy as np
y_point=np.array([2,3,4,5])
mp.plot(y_point,marker='*',ms=20) #ms=marker size
mp.show()
```

```python
#changing the edge color of marker
import matplotlib.pyplot as mp
import numpy as np
y_point=np.array([2,3,4,5])
mp.plot(y_point,marker='*',mec="red",ms=10) #mec=marker edge color but it have border color same as line
mp.show()
```

```python
#changing the face color of marker
import matplotlib.pyplot as mp
import numpy as np
y_point=np.array([2,3,4,5])
mp.plot(y_point,marker='*',mfc="red",markersize=20) #mfc=marker face color 
mp.show()
```

```python
#changing the line design as line linestyle and changing its color also
import matplotlib.pyplot as mp
import numpy as np
y_point=np.array([2,3,4,5])

#changing the color of line and its design
mp.plot(y_point,linestyle='dotted',color='blue',marker='*',mfc='red',ms=20) 
mp.show()
```

```python
#changing the line width in line design

#changing the line design as line linestyle and changing its color also
import matplotlib.pyplot as mp
import numpy as np
y_point=np.array([2,3,4,5])

#changing the line width in line
mp.plot(y_point,linestyle='dotted',linewidth=10,color='blue',marker='*') 
mp.show()
```

```python
#How to plot multiple graph in single graph set
import matplotlib.pyplot as plb
import numpy as np

#plot1
x1_point = np.array([0,6,4])
y1_point = np.array([50,20,6])


plb.subplot(1,2,1) #row,column,plotnumber
plb.plot(x1_point,y1_point)

#plot2
x1_point = np.array([2,0,6])
y1_point = np.array([3,50,20])

plb.subplot(1,2,2) #row,column,plotnumber
plb.plot(x1_point,y1_point)

#displayint all plot at once
plb.show()
```

```python
import matplotlib.pyplot as plt

# Data for the bar graph
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
values = [10, 20, 15, 25]
c=['blue', 'green', 'red', 'purple']

# Create a bar graph with custom colors and width
plt.bar(categories, values,color=c, width=0.5)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('colors')

# Show the plot
plt.show()
```
