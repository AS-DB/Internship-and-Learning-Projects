1.Linear Regression:-
It refers to established relationship between dependent and independent variable.Means finding the value of dependent variable.

                   e.g.

                  1. independent variable x= [1 2 3 4 5 6]
                   dependent variable y =x+1 then the value of y will be [2,3,4,5,6,7]
                  
                  2.x(week) - [1,2,3,4,5]
                    y(sales inn thousand) - [1.2,1.8,2.6,3.2,3.8]

-->To use linear regression in python we use sklearn.linear_model import LinearRegression

-->to install we use command pip install scikit-learn

We can directly undertand about it by seeing the below code 

```python
#question related linear regression

'''
Q1.Find the sales of 7th week
x(week) - [1,2,3,4,5]
y(sales inn thousand) - [1.2,1.8,2.6,3.2,3.8]
'''            

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


x=np.array([1,2,3,4,5]).reshape(-1,1)#as sklibrary needs 2d thats why we reshape it
y=np.array([1.2,1.8,2.6,3.2,3.8])

#fitting the model
model=LinearRegression()
model.fit(x,y)

#future value predication by adding elements
x_new =np.array([[7],[10],[12]])#As skllibrary needs 2d array
y_pred=model.predict(x_new)

print(f'Predicated sales for the 7th week:{y_pred}')

#plotting the graph
plt.scatter(x,y,color="red",label="Actual Sales")
plt.plot(x,model.predict(x),color='blue',label="Regression line")
plt.scatter(x_new,y_pred,color="green",label="predicated sales for 7th week")

#providing the labels in graphs
plt.xlabel("week")
plt.ylabel("sales")

plt.legend() #it will show the line,plot,prediccated value area with labeled marked 
plt.show()
```

2.Logistic Regression:-
Its a type of regression used to solve the classification problem.It will give output between 0 and 1.

e.g.
Q1.
Study hours:-    pass(1)/fail(0)
29                  0
15                  0
33                  1
28                  1
39                  1

How much percentage of chances that a student gone a pass of fail if he study 27 hour

```python
'''
Q1.
Study hours:-    pass(1)/fail(0)
29                  0
15                  0
33                  1
28                  1
39                  1

How much percentage of chances that a student gone a pass of fail if he study 33 hour
'''
#error in code for display

import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression

x=np.array([29,15,33,28,39]).reshape(-1,1)
y=np.array([0,0,1,1,1])

#fitting the model
model= LogisticRegression()
model.fit(x,y)

#NOw predict the output for it if he study for 33 hour
x_new=np.array([[33]])
y_prob=model.predict_proba(x_new)

#showing the output before graph
print(f"value is:{y_prob[0][1]:.2f}")

#ploting of points
x_values=np.linspace(10,45,100).reshape(-1,1)
y_values=model.predict_proba(x_values)[:, 1]

#ploltting of graphs
plt.scatter(x,y,color="blue",label="Actual value")
plt.plot(x_values,y_values,color="red",label="Logistic Regression Curve")
plt.scatter(x_new, y_prob[0][1], color="green", label="Probability for 33 hours")


plt.xlabel("Hours studied")
plt.ylabel("Probability of passing")

plt.legend()
plt.show()
```

```python
#chatgpt code for same logistic regression problem
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression

# Data
x = np.array([29, 15, 33, 28, 39]).reshape(-1, 1)
y = np.array([0, 0, 1, 1, 1])

# Fitting the model
model = LogisticRegression()
model.fit(x, y)

# Predicting the probability for a student who studies for 33 hours
x_new = np.array([[33]])
y_prob = model.predict_proba(x_new)

# Showing the output before graph
print(f"Probability of passing for 33 hours of study is: {y_prob[0][1]:.2f}")

x_values = np.linspace(10, 45, 100).reshape(-1, 1)
y_values = model.predict_proba(x_values)[:, 1]

# Plotting the graph
plt.plot(x_values, y_values, color="red", label="Logistic Regression Curve")
plt.scatter(x, y, color="blue", label="Training data")
plt.scatter(x_new, y_prob[0][1], color="green", label="Probability for 33 hours")

plt.xlabel("Hours studied")
plt.ylabel("Probability of passing")
plt.legend()
plt.show()
```
